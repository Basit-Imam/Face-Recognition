```
import face_recognition #access to face detection model in DLIB
```

```
img = face_recognition.load_image_file("img.jpg")
```

```
# Find all the faces in image
face_locations = face_recognition.face_locations(img)
number_of_faces = len(face_locations)
print("I found {} face(s) in this photograph.".format(number_of_faces))
```

```
#Load the image into Python Image Library object
pil_image = PIL.Image.fromarray(img)
```

```
for face_location in face_locations:
  top, right, bottom, left = face_location
  #print the location of each face in this image. Each face is a list of coordinates.
  print("A face is located at pixel location Top:{}, Left:{}, Bottom:{}, Right:{}".format(top,left, bottom, right))

  #Draw a box around the face
  draw = PIL.ImageDraw.Draw(pil_image)
  draw.rectangle([left, top, right, bottom], outline = 'red')
```
